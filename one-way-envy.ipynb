```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#define a function that checks if the direction of envy from 
#A = (x,y) and B = (t[0],t[1]) is in one direction only (i.e., from A to B)
def onewayenvy(x,y,t):
    p = t[0]
    q = t[1]
    
    a1 = np.sqrt((p-1)**2 + q**2)
    a2 = np.sqrt((p+1)**2 + q**2)
    
    b1 = np.sqrt((x-1)**2 + y**2)
    b2 = np.sqrt((x+1)**2 + y**2)
    
    ab = np.sqrt((p-x)**2 + (q-y)**2)
    
    envy_atob = np.heaviside(a2 - ab - b1, 0)
    no_envy_btoa = np.heaviside(ab + a1 - b2, 0)
    
    return envy_atob * no_envy_btoa
```

```python
#INPUT: a reference point A = (x,y) with both -2 < x < 2 and -2 < y < 2
#OUTPUT: A plot of all points B such that A envies B, but B does not envy A.
#NOTE: Centers are at (-1,0) and (1,0).

#Creating a grid 
x = np.linspace(-2, 2, 4000) 
y = np.linspace(-2, 2, 4000) 
x_1, y_1 = np.meshgrid(x, y) 


#Take input of the reference point
x = float(input("Enter x coordinate -2<x<2: "))
y = float(input("Enter y coordinate -2<y<2: "))
a = np.array([x,y])

#Compute the envy function defined above
z = onewayenvy(x_1,y_1,a) 

#plot figure
fig = plt.figure(figsize=(20,20))
ax = fig.gca()
ax.set_xticks(np.arange(-2, 2, 0.1))
ax.set_yticks(np.arange(-2, 2, 0.1))
plt.contourf(x_1, y_1, z, cmap = "binary")
plt.scatter(x,y, s = 200, marker = "x")
plt.scatter([-1,1],[0,0], s = 400)
plt.grid()
plt.show() 
```
